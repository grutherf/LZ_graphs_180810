```python
"""
The current way of creating this graph requires 13 versions of testNEST.cpp and thus 13 wrappers.
Each testNEST file has different values in nuisParams, varying Ly and Qy by +/- 10 and 20%
The naming convention is normal_name_of_file + what_is_varying + how_it_varies
For example, in the testNEST with both Qy and Ly increasing by 20%, we have:
testNEST + Both+p20 = testNESTBothp20.cpp

"""
#import wrappers
import nest_py_interfaceBothp10
import nest_py_interfaceBothm10
import nest_py_interfaceBothp20
import nest_py_interfaceBothm20

import nest_py_interfaceLyp10
import nest_py_interfaceLym10
import nest_py_interfaceLyp20
import nest_py_interfaceLym20

import nest_py_interfaceQyp10
import nest_py_interfaceQym10
import nest_py_interfaceQyp20
import nest_py_interfaceQym20
import nest_py_interface

import numpy as np

#get wrapper objects
npwBothp10 = nest_py_interfaceBothp10.nest_py_wrapper_std()
npwBothm10 = nest_py_interfaceBothm10.nest_py_wrapper_std()
npwBothp20 = nest_py_interfaceBothp20.nest_py_wrapper_std()
npwBothm20 = nest_py_interfaceBothm20.nest_py_wrapper_std()

npwLyp10 = nest_py_interfaceLyp10.nest_py_wrapper_std()
npwLym10 = nest_py_interfaceLym10.nest_py_wrapper_std()
npwLyp20 = nest_py_interfaceLyp20.nest_py_wrapper_std()
npwLym20 = nest_py_interfaceLym20.nest_py_wrapper_std()

npwQyp10 = nest_py_interfaceQyp10.nest_py_wrapper_std()
npwQym10 = nest_py_interfaceQym10.nest_py_wrapper_std()
npwQyp20 = nest_py_interfaceQyp20.nest_py_wrapper_std()
npwQym20 = nest_py_interfaceQym20.nest_py_wrapper_std()

npw = nest_py_interface.nest_py_wrapper_std()

#get results
numSims = 100
exposure = 5600 * 1000 * numSims

result = npw.testNEST([str(exposure), '8B','0','4','310','-1','1'])

resultBothp10 = npwBothp10.testNEST([str(exposure), '8B','0','4','310','-1','1'])
resultBothm10 = npwBothm10.testNEST([str(exposure), '8B','0','4','310','-1','1'])
resultBothp20 = npwBothp20.testNEST([str(exposure), '8B','0','4','310','-1','1'])
resultBothm20 = npwBothm20.testNEST([str(exposure), '8B','0','4','310','-1','1'])

resultLyp10 = npwLyp10.testNEST([str(exposure), '8B','0','4','310','-1','1'])
resultLym10 = npwLym10.testNEST([str(exposure), '8B','0','4','310','-1','1'])
resultLyp20 = npwLyp20.testNEST([str(exposure), '8B','0','4','310','-1','1'])
resultLym20 = npwLym20.testNEST([str(exposure), '8B','0','4','310','-1','1'])


resultQyp10 = npwQyp10.testNEST([str(exposure), '8B','0','4','310','-1','1'])
resultQym10 = npwQym10.testNEST([str(exposure), '8B','0','4','310','-1','1'])
resultQyp20 = npwQyp20.testNEST([str(exposure), '8B','0','4','310','-1','1'])
resultQym20 = npwQym20.testNEST([str(exposure), '8B','0','4','310','-1','1'])
```

```python
import matplotlib.pyplot as plt

#apply cuts
numEvAfterCuts = len(np.intersect1d(np.where(np.absolute(result.s1_n_hits) >=3) ,
                                    np.where(np.absolute(result.s2_raw_area_phd) >= 415)))/numSims

numEvAfterCutsQyp10 = len(np.intersect1d(np.where(np.absolute(resultQyp10.s1_n_hits) >=3) ,
                                         np.where(np.absolute(resultQyp10.s2_raw_area_phd) >= 415)))/numSims
numEvAfterCutsQyp20 = len(np.intersect1d(np.where(np.absolute(resultQyp20.s1_n_hits) >=3) ,
                                         np.where(np.absolute(resultQyp20.s2_raw_area_phd) >= 415)))/numSims
numEvAfterCutsQym10 = len(np.intersect1d(np.where(np.absolute(resultQym10.s1_n_hits) >=3) ,
                                         np.where(np.absolute(resultQym10.s2_raw_area_phd) >= 415)))/numSims
numEvAfterCutsQym20 = len(np.intersect1d(np.where(np.absolute(resultQym20.s1_n_hits) >=3) ,
                                         np.where(np.absolute(resultQym20.s2_raw_area_phd) >= 415)))/numSims

numEvAfterCutsLyp10 = len(np.intersect1d(np.where(np.absolute(resultLyp10.s1_n_hits) >=3) ,
                                         np.where(np.absolute(resultLyp10.s2_raw_area_phd) >= 415)))/numSims
numEvAfterCutsLyp20 = len(np.intersect1d(np.where(np.absolute(resultLyp20.s1_n_hits) >=3) ,
                                         np.where(np.absolute(resultLyp20.s2_raw_area_phd) >= 415)))/numSims
numEvAfterCutsLym10 = len(np.intersect1d(np.where(np.absolute(resultLym10.s1_n_hits) >=3) ,
                                         np.where(np.absolute(resultLym10.s2_raw_area_phd) >= 415)))/numSims
numEvAfterCutsLym20 = len(np.intersect1d(np.where(np.absolute(resultLym20.s1_n_hits) >=3) ,
                                         np.where(np.absolute(resultLym20.s2_raw_area_phd) >= 415)))/numSims

numEvAfterCutsBothp10 = len(np.intersect1d(np.where(np.absolute(resultBothp10.s1_n_hits) >=3) ,
                                           np.where(np.absolute(resultBothp10.s2_raw_area_phd) >= 415)))/numSims
numEvAfterCutsBothp20 = len(np.intersect1d(np.where(np.absolute(resultBothp20.s1_n_hits) >=3) ,
                                           np.where(np.absolute(resultBothp20.s2_raw_area_phd) >= 415)))/numSims
numEvAfterCutsBothm10 = len(np.intersect1d(np.where(np.absolute(resultBothm10.s1_n_hits) >=3) ,
                                           np.where(np.absolute(resultBothm10.s2_raw_area_phd) >= 415)))/numSims
numEvAfterCutsBothm20 = len(np.intersect1d(np.where(np.absolute(resultBothm20.s1_n_hits) >=3) ,
                                           np.where(np.absolute(resultBothm20.s2_raw_area_phd) >= 415)))/numSims

scaling = [.8, .9, 1, 1.1, 1.2]

Qyevents = np.array([numEvAfterCutsQym20,numEvAfterCutsQym10,numEvAfterCuts,numEvAfterCutsQyp10,numEvAfterCutsQyp20])

Lyevents = np.array([numEvAfterCutsLym20,numEvAfterCutsLym10,numEvAfterCuts,numEvAfterCutsLyp10,numEvAfterCutsLyp20])

Bothevents = np.array([numEvAfterCutsBothm20,numEvAfterCutsBothm10,numEvAfterCuts,numEvAfterCutsBothp10,numEvAfterCutsBothp20])

#print(Qyevents)
#print(Lyevents)
#print(Bothevents)

plt.plot(scaling,Qyevents, label = "Charge Yield")
plt.plot(scaling,Lyevents, label = "Light Yield")
plt.plot(scaling,Bothevents, label = "Both Yields")

Qyerr = np.sqrt(Qyevents*numSims)/numSims
Lyerr = np.sqrt(Lyevents*numSims)/numSims
Botherr = np.sqrt(Bothevents*numSims)/numSims

plt.errorbar(scaling, Qyevents, yerr = Qyerr, fmt = '.')
plt.errorbar(scaling, Lyevents, yerr = Lyerr, fmt = '.')
plt.errorbar(scaling, Bothevents, yerr = Botherr, fmt = '.')

#plt.plot(scaling,Qyevents)
plt.xlabel('yield scaling')
plt.ylabel('8B events passing cut')
plt.yscale('log')
plt.xscale('log')
plt.xlim([.75,1.25])
plt.xticks([.8,.9,1,1.1,1.2])
plt.ylim([min(Bothevents)-5,max(Bothevents)+20])
plt.title('NR Charge Yield Scaling vs 8B Event Rates (raw signals)')
plt.grid(True,which='both',linewidth=.5)
plt.legend()
plt.show()
```
